```python
from bs4 import BeautifulSoup
import urllib.request
import time
from tqdm import tqdm_notebook
import pandas as pd
import re
```

```python
wiki_network_directory = '../data/wiki_network'
```

```python
def to_df(seed, links): #Function to convert to dictionary of lists to datframe
    
    links_df = pd.DataFrame()
    _prev = seed.split('.org')[1]
    
    #print(_prev)
    
    for depth in links:
        for link in links[depth]:
            links_df = links_df.append({'depth': depth,
                                        'link': link,
                                        'prev': _prev}, ignore_index = True)
            
    return links_df


def start(seed): #Start the crawl function and save the results to file

    links        = {0:[seed]}
    max_recDepth = 6
    max_count    = 1000
    count        = 0
    recDepth     = 0
    
    while (count < max_count) and recDepth < max_recDepth:

        links, recDepth, count = crawl(links, recDepth, count, max_count)

        
    links_df = to_df(seed, links)
    links_df.to_csv('../data/wiki_network/' + seed.split('/wiki/')[1] + '.csv', index = False)
    
    return links


def crawl(links, recDepth = 0, count = 0, max_count = 1000, write_html = False): #Perform the actual crawl
   
        recDepth += 1
        links.update({recDepth:['/wiki/']})
        
        for link in tqdm_notebook(links[recDepth - 1]): 

            print ('\n\n')
            print (link)
            

            time.sleep(1)

            if 'https://en.wikipedia.org' not in link:
                link = 'https://en.wikipedia.org' + link

            resp = urllib.request.urlopen(link)
            soup = BeautifulSoup(resp, from_encoding=resp.info().get_param('charset'))
            
            if write_html:
                with open("../data/" + link.split('/wiki/')[1] + "html", "w") as file:
                    file.write(str(soup.prettify("utf-8")))

            for bc in soup.find_all(id="bodyContent"):

                for link in soup.body.find_all('a', href=True):

                    if (':' not in link['href'] and \
                        '//' not in link['href'] and\
                        '/Main_Page' not in link['href']) and ('/wiki/' in link['href']) and\
                        link['href'].split('#')[0] not in links[recDepth]:  #Ensuring that the link is a valid one
                            
                            count += 1
                            links[recDepth].append(link['href'].split('#')[0])
                            
                            print(str(link['href'].split('#')[0]) + ' Count: ' + str(count) + links[recDepth][-1]) 
                    
                    if count > max_count:
                        return links, recDepth, count 


        #links[recDepth + 1] = list(set(links[recDepth + 1]))
        
        return links, recDepth, count
```

## Task 3: Focused Crawling

```python
def focussed_start(keyword, seed): #Focussed crawl with keywords 

    links        = {0:[seed]}
    max_recDepth = 6
    max_count    = 1000
    count        = 0
    recDepth     = 0
    
    while (count < max_count) and recDepth < max_recDepth:

        links, recDepth, count = focused_crawl(keyword, links, recDepth, count, max_count)

        
    links_df = to_df(links)
    links_df.to_csv('data/' + seed.split('/wiki/')[1] + '.csv', index = False)
    
    return links

def focused_crawl(keyword, links, recDepth = 0, count = 0, max_count = 1000):
   
        recDepth += 1
        links.update({recDepth:['/wiki/']})
        all_links = []
        {all_links.extend(val) for depth, val in links.items()}
        
        
        for link in tqdm_notebook(links[recDepth - 1]): 

            print ('\n\n')
            print (link)
            
            time.sleep(1)

            if 'https://en.wikipedia.org' not in link:
                link = 'https://en.wikipedia.org' + link

            resp = urllib.request.urlopen(link)
            soup = BeautifulSoup(resp, from_encoding=resp.info().get_param('charset'))
            
            if re.findall((r"\b" + keyword + r"\b"), str(soup.prettify("utf-8"))):
                
                with open("data/" + link.split('/wiki/')[1].replace('/', '_') + ".html", "w") as file:
                    file.write(str(soup.prettify("utf-8")))

                for bc in soup.find_all(id="bodyContent"):

                    for link in soup.body.find_all('a', href=True):

                        if (':' not in link['href'] and \
                            '//' not in link['href'] and \
                            '/Main_Page' not in link['href']) and ('/wiki/' in link['href']) and\
                            link['href'].split('#')[0] not in all_links:

                            if re.findall((r"\b" + keyword + r"\b"), str(link.string), flags=re.IGNORECASE) or\
                                re.findall((r"." + keyword + r"."), str(link['href']), flags=re.IGNORECASE):
           
                                count += 1
                                links[recDepth].append(link['href'].split('#')[0])
                                all_links.append(link['href'].split('#')[0])

                                print(str(link['href'].split('#')[0]) + ' Count: ' + str(count) + links[recDepth][-1]) 

                        if (count > max_count):
                            return links, recDepth, count 


        #links[recDepth + 1] = list(set(links[recDepth + 1]))
        
        return links, recDepth, count
```
